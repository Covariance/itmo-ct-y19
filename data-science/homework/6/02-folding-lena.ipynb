# Свёртка Лены

```python
import cv2
import torch
import torchvision
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set_style()
```

Скачаем себе Лену:

Нарисуем Лену!

```python
img = cv2.cvtColor(cv2.imread('pics/lena_inf.jpg'), cv2.COLOR_BGR2RGB)

plt.figure(figsize=(6, 6))
plt.title('Mathematically Correct Lena')
plt.imshow(img)
plt.axis('off')
plt.show()
```

Выберем какой-нибудь алгоритм для edge detection [отсюда](https://www.theobjects.com/dragonfly/dfhelp/3-5/Content/05_Image%20Processing/Edge%20Detection%20Filters.htm). Я решил остановиться на Prewitt.

```python
kernelx = np.array([[1, 0],[0, -1]])
kernely = np.array([[0, 1],[-1, 0]])
img_roberts_cross_x = cv2.filter2D(img, -1, kernelx)
img_roberts_cross_y = cv2.filter2D(img, -1, kernely)
img_roberts_cross = img_roberts_cross_x + img_roberts_cross_y

plt.figure(figsize=(5, 5))
plt.title('Mathematically Correct Lena With Prewitt Filter')
plt.imshow(img_roberts_cross)
plt.axis('off')
plt.show()
```
